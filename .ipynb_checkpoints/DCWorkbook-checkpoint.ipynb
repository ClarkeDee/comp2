```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# instantiate the model (using the default parameters)
logreg = LogisticRegression()
```

```python
#columns we need to divide into target variables and 
#'ID', 'LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_0',
 #      'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2',
  #     'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1',
   #    'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6',
    #   'default payment next month'
```

```python
# adding variables for x and y, y is our target variable
x= data[['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_0','PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2','BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']]
y = data['default payment next month']
```

```python
features = x
target= y
```

```python
logreg.fit(x,y)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# 30 / 70
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=.30, random_state=2019)

len(x_train)
len(y_train)
```

```python
len(x_test)
len(y_test)
```

```python

# fit the model with data
logreg.fit(x_train,y_train)

#
y_pred=logreg.predict(x_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#generating model using svm
```

```python
x1= data[['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_0','PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2','BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']]
y1 = data['default payment next month']
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# 30 / 70
x1_train, x1_test, y1_train, y1_test= train_test_split(x1,y1, test_size=.30, random_state=2019)
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(x1_train, y1_train)

#Predict the response for test dataset
y_pred2 = clf.predict(x1_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred2))
```
